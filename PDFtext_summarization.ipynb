```python
# Install necessary libraries
!pip install PyPDF2
!pip install transformers

# Import required libraries
from PyPDF2 import PdfReader
from transformers import pipeline

# Load the summarization pipeline
def load_summarizer():
    print("Loading summarization model...")
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    print("Model loaded successfully!")
    return summarizer

summarizer = load_summarizer()

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    print("Extracting text from the PDF...")
    pdf_reader = PdfReader(pdf_path)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text()
    print("Text extraction complete!")
    return text

# Function to summarize text
def summarize_text(text, max_length=200, min_length=50):
    print("Summarizing the text...")
    summary = summarizer(text[:1024], max_length=max_length, min_length=min_length, do_sample=False)[0]['summary_text']
    print("Summary generated successfully!")
    return summary

# Upload and process the PDF file
from google.colab import files

print("Please upload a research paper (PDF).")
uploaded_file = files.upload()

if uploaded_file:
    # Get the file name
    file_name = list(uploaded_file.keys())[0]

    # Extract text from the PDF
    paper_text = extract_text_from_pdf(file_name)

    if paper_text.strip():
        # Summarize the extracted text
        summary = summarize_text(paper_text)

        # Display the results
        print("\n--- Summary of the Research Paper ---\n")
        print(summary)
    else:
        print("No text could be extracted from the PDF. Please ensure it's a valid and readable PDF.")
```
